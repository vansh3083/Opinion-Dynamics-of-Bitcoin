## Lemmatization

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

# Download NLTK resources (run once)
# nltk.download('punkt')
# nltk.download('wordnet')

# Initialize the WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

# Define a function to lemmatize text
def lemmatize_text(text):
    # Tokenize the text
    tokens = word_tokenize(text)
    # Lemmatize each token
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
    # Join the lemmatized tokens into a single string
    lemmatized_text = ' '.join(lemmatized_tokens)
    return lemmatized_text

# Apply the function to lemmatize the 'full_text' column
eth_data['lemmatized_text'] = eth_data['full_text'].apply(lemmatize_text)

# Display the first few rows of the DataFrame to verify the changes
print(eth_data[['full_text', 'lemmatized_text']].head())
```

## Sentiment Analysis using VADER

```python
import nltk
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Initialize VADER sentiment analyzer
sid = SentimentIntensityAnalyzer()

# Function to get sentiment label based on polarity score
def get_sentiment_label(score):
    if score > 0:
        return 1  # Positive sentiment
    elif score < 0:
        return -1  # Negative sentiment
    else:
        return 0  # Neutral sentiment

# Function to analyze sentiment of a text and return sentiment label
def analyze_sentiment(text):
    # Get polarity scores from VADER
    sentiment_scores = sid.polarity_scores(text)
    # Get compound score which represents overall sentiment
    compound_score = sentiment_scores['compound']
    # Get sentiment label based on compound score
    sentiment_label = get_sentiment_label(compound_score)
    return sentiment_label

# Apply sentiment analysis to each tweet
eth_data['sentiment'] = eth_data['lemmatized_text'].apply(analyze_sentiment)

# Display the first few rows of the DataFrame with sentiment labels
print(eth_data[['lemmatized_text', 'sentiment']].head())
```

## Topic Modeling using LDA

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import numpy as np

# Tokenize the text and remove stopwords
# You may need to perform additional preprocessing steps here, such as lemmatization
# Define a CountVectorizer with suitable parameters for tokenization and preprocessing
vectorizer = CountVectorizer(max_df=0.95, min_df=2, stop_words='english')

# Fit and transform the text data to create a document-term matrix
dtm = vectorizer.fit_transform(eth_data['lemmatized_text'])

# Initialize LDA with the desired number of topics
num_topics = 5  # Adjust the number of topics as needed
lda = LatentDirichletAllocation(n_components=num_topics, random_state=42)

# Fit LDA to the document-term matrix
lda.fit(dtm)

# Display the top words for each topic
feature_names = vectorizer.get_feature_names_out()
for topic_idx, topic in enumerate(lda.components_):
    print(f"Topic {topic_idx + 1}:")
    top_words_idx = topic.argsort()[:-10-1:-1]  # Get indices of top 10 words for each topic
    top_words = [feature_names[i] for i in top_words_idx]
    print(top_words)

# Transform documents into topic distributions
topic_distributions = lda.transform(dtm)

# Assign each document to its most probable topic
eth_data['top_topic'] = np.argmax(topic_distributions, axis=1)
```

## Topics Dominance over Time

```python
# Assuming you have already performed LDA and obtained topic distributions for each document

# Concatenate topic distributions with the date information
topic_distributions_with_date = pd.concat([eth_data['date'], pd.DataFrame(topic_distributions)], axis=1)

# Group by date and calculate the average topic distribution for each date
average_topic_distribution_by_date = topic_distributions_with_date.groupby('date').mean()

# Get the dominant topic for each date
dominant_topic_by_date = average_topic_distribution_by_date.idxmax(axis=1)

# Plot the prevalence of dominant topics over time
plt.figure(figsize=(10, 6))
plt.plot(dominant_topic_by_date.index, dominant_topic_by_date.values, marker='o', linestyle='-')
plt.xlabel('Date')
plt.ylabel('Dominant Topic')
plt.title('Dominant Topic Over Time')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Named Entity Recognition

```python
import spacy
import pandas as pd

# Load the spaCy model with NER support
nlp = spacy.load("en_core_web_sm")

# Function to perform NER on text and extract entities
def extract_entities(text):
    doc = nlp(text)
    entities = [(ent.text, ent.label_) for ent in doc.ents]
    return entities

# Apply NER to the lemmatized text column
eth_data['entities'] = eth_data['lemmatized_text'].apply(extract_entities)

# Display the DataFrame with extracted entities
print(eth_data[['lemmatized_text', 'entities']].head())
```

```python
import matplotlib.pyplot as plt

# Flatten the list of entities from the DataFrame and convert to strings
all_entities = [' '.join(entity) for entities_list in eth_data['entities'] for entity in entities_list]

# Count the occurrences of each entity
entity_counts = {}
for entity in all_entities:
    if len(entity) > 3:  # Filter entities with string length more than 3
        entity_counts[entity] = entity_counts.get(entity, 0) + 1

# Sort the entities by their occurrence in descending order
sorted_entities = sorted(entity_counts.items(), key=lambda x: x[1], reverse=True)

# Get the top 10 entities
top_entities = sorted_entities[:10]

# Extract entity labels and occurrence counts
entity_labels = [entity[0] for entity in top_entities]
occurrence_counts = [entity[1] for entity in top_entities]

# Plot the top 10 entities and their occurrence counts
plt.figure(figsize=(10, 6))
plt.barh(entity_labels, occurrence_counts, color='skyblue')
plt.xlabel('Occurrence Count')
plt.ylabel('Entity')
plt.title('Top 10 Entities by Occurrence Count (String Length > 3)')
plt.gca().invert_yaxis()  # Invert y-axis to display the most frequent entities at the top
plt.tight_layout()
plt.show()
```
